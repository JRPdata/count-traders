```python
def get_unique_traders_on_market(slug):
    # URL containing retrieve market from slug
    market_url = f"https://manifold.markets/api/v0/slug/{slug}"

    market = get_api(market_url)
    ids = set()
    if market:
        market_id = market['id']

        # URL containing position details
        positions_url = f"https://manifold.markets/api/v0/market/{market_id}/positions"
        
        positions = get_api(positions_url)
        if positions:
            for contract in positions:
                if 'hasShares' in contract and contract['hasShares']:
                    ids.add(contract['userId'])
    return ids

def get_unique_traders_for_markets(slugs):
    all_ids = set()
    i = 1
    n = len(slugs)
    for slug in slugs:
        print(f"({i} / {n}) Getting info for {slug}")
        ids = get_unique_traders_on_market(slug)
        all_ids = all_ids.union(ids)
        i += 1
    return all_ids
```
